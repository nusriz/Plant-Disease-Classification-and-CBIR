```
import os
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```
read_path = 'drive/My Drive/Dataset/Bacterial Spot'
X = []



for file in os.listdir(read_path):
	bgr_img = cv2.imread(os.path.join(read_path,file))	
	hist_b = np.asarray(cv2.calcHist([bgr_img],[0],None,[256],[0,256]))
	hist_g = np.asarray(cv2.calcHist([bgr_img],[1],None,[256],[0,256]))
	hist_r = np.asarray(cv2.calcHist([bgr_img],[2],None,[256],[0,256]))
	hist_b = hist_b.astype('uint8')
	hist_g = hist_g.astype('uint8')
	hist_r = hist_r.astype('uint8')
	hist_b = hist_b.T[0]
	hist_g = hist_g.T[0]
	hist_r = hist_r.T[0]
	hist = np.concatenate((hist_b,hist_g))
	hist = np.concatenate((hist,hist_r))
	X.append(hist)


features = np.asarray(X)
print(features.shape)
```

```
import cloudpickle as pickle
with open('drive/My Drive/Dataset/bspot.pkl', 'wb') as outfile:
    pickle.dump(features, outfile)
```
