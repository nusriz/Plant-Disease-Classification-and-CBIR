```
import csv
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

def queryFeatures():
  with open('drive/My Drive/Features/Feature_vector.csv') as f :
    read= csv.reader(f)
    c =0
    for row in read:
      if c==0:
        features=[float(x) for x in row[0:]]
      c=1

  return features
    


def chi2_distance(A,B):
  eps = 1e-5
  print(A)
  print(B)
  d = 0.5 * np.sum([((a - b) ** 2) / (a + b + eps)
  for (a, b) in zip(A, B)])

  return d


with open('drive/My Drive/Dataset/Feature_vector_imno.csv') as csv_file :
    csv_reader = csv.reader(csv_file)
    
    results={}
    queryfeatures=queryFeatures()
    for row in csv_reader:
      #print(row)
      features = [float(x) for x in row[1:]]
      d=chi2_distance(features,queryfeatures)
      #print(d)

      results[row[0]]=d
      

      #print(result)
    
    csv_file.close()

print(len(results))
print(results)
results = sorted([(v, k) for (k, v) in results.items()])
#print(results)
		# return our (limited) results
results=results[:10]
print(results)
```
